```python
# columns trimming

twitter = twitter.drop(twitter.columns[[1,3,4]], axis=1)
twitter.columns.values[0] = "sentiment"
twitter.columns.values[1] = "date"
twitter.columns.values[2] = "tweet"

twitter.head(5)
```

```python
# removing stopwords

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import TweetTokenizer

nltk.download('stopwords')
nltk.download('punkt')

stop_words_cached = stopwords.words('english')
tk = TweetTokenizer()


import re
WORD = re.compile(r"")
def regTokenize(text):
    words = WORD.findall(text)
    return words

tokenized_twitter = twitter.copy()
tokenized_twitter.head(5)

tokenized_twitter["tweet"] = tokenized_twitter["tweet"].apply(tk.tokenize)
tokenized_twitter.head(5)
```

```python
tokenized_twitter.head(1)
tokenized_twitter.to_csv("tokenized_twitter.csv")
```

```python
#eliminate @ handles? - removing stopwords

#extend stopwords?
tokenized_twitter.head(20)

def remove_stopwords(text_array):
    fresh_text_array = []
    for word in text_array:
      lowercase_word = word.lower()
      if lowercase_word not in stop_words_cached:
        fresh_text_array.append(lowercase_word)
    return fresh_text_array

no_stopwords_twitter = tokenized_twitter.copy()
no_stopwords_twitter["tweet"] = tokenized_twitter['tweet'].apply(remove_stopwords)

no_stopwords_twitter.head(10)
```

```python
from transformers import pipeline

pipe = pipeline("text-classification", model="finiteautomata/bertweet-base-sentiment-analysis", device=0)

def label_to_number(label):
  if label == 'NEG':
    return 0
  if label == 'POS':
    return 4
  if label == 'NEU':
    return 2

def create_tweet_list():
  for tweet in twitter["tweet"]:
    if len(tweet) > 128:
      tweet = tweet[:128]
    tweet_list.append(tweet)

def calculate_sentiment_with_outside_model():
  global BATCH_COUNT
  while BATCH_COUNT < (1600000 / BATCH_SIZE):
    print("Starting batch " + str(BATCH_COUNT+1))
    local_sentiment_calculated = pipe(tweet_list[(BATCH_SIZE * BATCH_COUNT):(BATCH_SIZE * (BATCH_COUNT + 1))])
    print("Completed batch " + str(BATCH_COUNT+1) + " processed " + str((BATCH_COUNT + 1) * BATCH_SIZE) + "/" + str(1600000) + "tweets." )
    for result in local_sentiment_calculated:
      sentiment_calculated.append(label_to_number(result['label']))
    BATCH_COUNT = BATCH_COUNT + 1

tweet_list = []

create_tweet_list()
calculate_sentiment_with_outside_model()
```
